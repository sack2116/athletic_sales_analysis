### 1. Combine and Clean the Data

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
df_combined = pd.concat([df_2020, df_2021], ignore_index=True)
```

```python
# Check if any values are null.
df_combined.isnull().sum()
```

```python
# Check the data type of each column
df_combined.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
# df_combined['invoice_date'] = pd.to_datetime(df_combined['invoice_date'])
df_combined['invoice_date'] = pd.to_datetime(df_combined['invoice_date'], format='mixed')
```

```python
# Confirm that the "invoice_date" data type has been changed.
df_combined.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
products_sold = df_combined.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()
# Rename the sum to "Total_Products_Sold".
products_sold.rename(columns={'units_sold': 'total_products_sold'}, inplace=True)
top_regions_products_sold = products_sold.sort_values(by='total_products_sold', ascending=False).head(5)

# Show the top 5 results.
top_regions_products_sold
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
products_sold_pivot = pd.pivot_table(df_2020, 
                                     values='units_sold', 
                                     index=['region', 'state', 'city'], 
                                     aggfunc='sum')
#sjust reseting indec
products_sold_pivot = products_sold_pivot.reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
products_sold_pivot.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
top_5_products_sold_pivot = products_sold_pivot.sort_values(by='Total_Products_Sold', ascending=False).head(5)
top_5_products_sold_pivot
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
sales_grouped = df_2020.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()
# Rename the "total_sales" column to "Total Sales"
sales_grouped.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
top_5_sales = sales_grouped.sort_values(by='Total Sales', ascending=False).head(5)
top_5_sales
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
sales_pivot = pd.pivot_table(df_2020, 
                             values='total_sales', 
                             index=['region', 'state', 'city'], 
                             aggfunc='sum')
#reseting index
sales_pivot = sales_pivot.reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
sales_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
top_5_sales_pivot = sales_pivot.sort_values(by='Total Sales', ascending=False).head(5)
top_5_sales_pivot
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
grouped_sales = df_2020.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].sum().reset_index()
# Rename the "total_sales" column to "Total Sales"
grouped_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
top_5Grouped_sales = grouped_sales.sort_values(by='Total Sales', ascending=False).head(5)
top_5Grouped_sales
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
sales_pivot = pd.pivot_table(df_2020, 
                             values='total_sales', 
                             index=['retailer', 'region', 'state', 'city'], 
                             aggfunc='sum')
#Just reseting the index
sales_pivot = sales_pivot.reset_index()
# Optional: Rename the "total_sales" column to "Total Sales"
sales_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
top_5_sales_pivot = sales_pivot.sort_values(by='Total Sales', ascending=False).head(5)      
top_5_sales_pivot
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear_sales = df_combined[df_combined['product'] == "Women's Athletic Footwear"]
womens_footwear_sales
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_footwear_grouped = womens_footwear_sales.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_grouped.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)
# Sorting it
top_5_womens_footwear = womens_footwear_grouped.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)
# Show the top 5 results.
top_5_womens_footwear
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_footwear_pivot = womens_footwear_sales.pivot_table(
    index=['retailer', 'region', 'state', 'city'],
    values='units_sold',
    aggfunc='sum'
).reset_index()


# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_pivot.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)
#sorting them in descending order
top_5_womens_footwear_pivot = womens_footwear_pivot.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)
# Show the top 5 results.
top_5_womens_footwear_pivot
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
sales_pivot = df_combined.pivot_table(
    index='invoice_date',
    values='total_sales',
    aggfunc='sum'
).reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
sales_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the table.
sales_pivot
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
df_combined['invoice_date'] = pd.to_datetime(df_combined['invoice_date'], format='mixed')
df_combined.set_index('invoice_date', inplace=True)
#print(df_combined.index)
daily_resampled_sales = df_combined.resample('D').sum()
daily_resampled_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)


# Sort the resampled pivot table in descending order on "Total Sales".
sorted_daily_sales = daily_resampled_sales.sort_values(by='Total Sales', ascending=False)
sorted_daily_sales.head()
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_resampled_sales = df_combined.resample('W').sum()
weekly_resampled_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
# Sort the resampled pivot table in descending order on "Total Sales".
sorted_weekly_sales = weekly_resampled_sales.sort_values(by='Total Sales', ascending=False)
```
